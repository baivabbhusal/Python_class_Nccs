```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

### Data Loading

```python
df=pd.read_csv("data/lr-Real-estate(1).csv")
df.head(2)
```

```python
df.columns
```

```python
df=df.drop(['No', 'X1 transaction date'],axis=1)
df.head(2)
```

```python
df=df.rename(
    columns={ 
        'X2 house age':'age',
       'X3 distance to the nearest MRT station':'dist-mrt station',
       'X4 number of convenience stores':'n stores', 
       'X5 latitude':'lat', 
       'X6 longitude':'lon',
       'Y house price of unit area':'price'
    }
)
df.head(2)
```

```python
df.shape #(num of records,num of features)
```

```python
# independent and dependent features split
x=df.drop('price',axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    x,
    y,
    test_size=0.2) #or train_size=0.8
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    x,
    y,
    test_size=0.2, )#or train_size=0.8
    # random_states=42
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
x_train.head(3)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression(n_jobs=-1)

lin_reg.fit(x_train,y_train)
```

```python
y_pred=lin_reg.predict(x_test)
```

```python
temp=x_test.copy()
temp['y_test']=y_test
temp['y_pred']=y_pred
temp.head(10)
```

```python
from sklearn.metrics import(
    mean_squared_error,
    root_mean_squared_error,
    r2_score
)
```

```python
print(f"{mean_squared_error(y_test,y_pred)=}")
print(f"{root_mean_squared_error(y_test,y_pred)=}")
print(f"{r2_score(y_test,y_pred)=}")
```

```python
import joblib 
joblib.dump(
    lin_reg,
    './models.lin_reg_model_1.joblib'
)
```
