```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# can use list/tuple/numpy array/pandas series
x=np.array([2,3,4,5,6,9,11])
y=np.array([3,5,4,9,2,5,3])
title_font={"family":"serif",}
plt.figure(figsize=(5,3)) #figsize=(width,height)
plt.plot(x,y,color="red",label='A')
plt.plot(x*2,y-2,color="green",label='B')
plt.title("Student's Flow in nccs")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
#plt.grid(axis='x') for vertical line only
plt.grid() #or plt.grid(axis='both)
plt.legend() #for specifying line 
```
