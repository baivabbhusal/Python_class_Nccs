```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# can use list/tuple/numpy array/pandas series
x=np.array([2,3,4,5,6,9,11])
y=np.array([3,5,4,9,2,5,3])
title_font={"family":"serif","color":"red","size":20}
label_font={"family":"serif","color":"blue","size":20}
plt.figure(figsize=(5,3)) #figsize=(width,height)
plt.plot(x,y,color="red",label='A',marker='*')
plt.plot(x*2,y-2,color="green",label='B',marker='<')
plt.title("Student's Flow in nccs",fontdict=title_font)
plt.xlabel("X-Label",fontdict=label_font)
plt.ylabel("Y-Label",fontdict=label_font)
#plt.grid(axis='x') for vertical line only
plt.grid() #or plt.grid(axis='both)
plt.legend() #for specifying line 
```

```python
x=np.array([2,3,4,5,6,9,11])
y=np.array([3,5,4,9,2,5,3])
plt.figure(figsize=(5,2))
plt.plot(x,y,marker='*')
```

```python
plt.figure(figsize=(5,2))
plt.plot(x,y,marker='*',ms=15,mec='red',mfc='green')
```

```python
plt.figure(figsize=(5,2))
plt.plot(x,y,'*')
```

```python
plt.figure(figsize=(5,2))
plt.plot(x,y,">:g")
plt.plot(x,y+1,"*--b")
plt.plot(x,y+2,"^-r")#marker border-type color
plt.plot(x,y+3,"d-.y",linewidth=3,ms=15)
```

```python
plt.figure(figsize=(5,2))
plt.scatter(x,y,color='green',marker='>')
```

### Bar Graph

```python
labels=['car','bus','truck','bicycle']
price_in_k=[1003,3005,4001,505]
plt.bar(labels,price_in_k,color='green',width=0.5,align='edge')
```

```python
plt.figure(figsize=(5,2))
plt.barh(labels,price_in_k,height=0.4)
```

```python
price_in_k=np.array([1002,2004,2055,505])
plt.figure(figsize=(5,2))
plt.barh(labels,price_in_k,height=0.4)
plt.barh(labels,price_in_k+400)
```

### Pie Chart

```python
price_in_k=np.array([1002,2004,2055,505])
plt.figure(figsize=(5,2))
plt.pie(price_in_k,labels=labels)
plt.tight_layout()
```

```python
price_in_k=np.array([1002,2004,2055,505])
plt.figure(figsize=(5,2))
explodes=[0,0,0,0.3]
plt.pie(price_in_k,labels=labels,explode=explodes)
plt.tight_layout()
```

```python
nr,nc=2,2
plt.subplot(nr,nc,1)
explodes=[0,0.5,0,0]
plt.pie(price_in_k,labels=labels,explode=explodes)
plt.tight_layout()
```

```python
nr,nc=2,2
plt.subplot(nr,nc,1)
explodes=[0,0.5,0,0]
plt.pie(price_in_k,labels=labels,explode=explodes)
plt.tight_layout()

plt.subplot(nr,nc,4)
explodes=[0.2,0,0,0]
plt.pie(price_in_k,labels=labels,explode=explodes)
plt.tight_layout()
```
