```python
df.drop(["Pregnancies","BloodPressure","SkinThickness"],axis=1,inplace=True)
```

```python
# Features
x = df.drop("Outcome", axis=1)  # drop target column

# Target
y = df["Outcome"] 
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
y_pred = log_reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
print(f"{accuracy_score(y_test,y_pred)=}")
print(f"{recall_score(y_test,y_pred)=}")
print(f"{precision_score(y_test,y_pred)=}")
print(f"{f1_score(y_test,y_pred)=}")
```
